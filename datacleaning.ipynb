```python
data["Weight"] = data["Weight"].str[:3].astype(float) /2.205
```

```python
data['HeightCM'] = data['Height'].str.split("'")
data['feet'] = data['HeightCM'].str.get(0).astype(float)
data['inch'] = data['HeightCM'].str.get(1).astype(float)
data['Height'] = (data['feet']*12+data['inch'])*2.54
```

```python
c1=data.loc[:,['Name']]
c2=data.loc[:,['Position']]
```

```python
data.dropna(how='any',subset=['Name','Position'], inplace=True)
```
